## feature engineering

```python
import pandas as pd
import numpy as np
import xgboost as xgb

np.random.seed(2018)

trn = pd.read_csv('kaggle/input/santander-product-recommendation-data/train_ver2.csv')
tst = pd.read_csv('kaggle/input/santander-product-recommendation-data/test_ver2.csv')

prods = trn.columns[24:].tolist()
trn[prods] = trn[prods].fillna(0.0).astype(np.int8)

no_product = trn[prods].sum(axis=1) == 0
trn = trn[~no_product]

for col in trn.columns[24:]:
    tst[col] = 0
df = pd.concat([trn, tst], axis=0)

features = []

categorical_cols = ['ind_empleado', 'pais_residencia','sexo','tiprel_1mes', 'indresi', 'indext', 'conyuemp','canal_entrada','indfall','tipodom','nomprov','segmento']
for col in categorical_cols:
    df[col],_= df[col].factorize(na_sentinel=-99)
features+=categorical_cols

df['age'].replace(' NA', -99, inplace=True)
df['age'] = df['age'].astype(np.int8)

df['antiguedad'].replace('     NA', -99, inplace=True)
df['antiguedad'] = df['antiguedad'].astype(np.int8)

df['renta'].replace('         NA', -99, inplace=True)
df['renta'].fillna(-99, inplace=True)
df['renta'] = df['renta'].astype(float).astype(np.int8)

df['indrel_1mes'].replace('P', 5, inplace=True)
df['indrel_1mes'].fillna(-99, inplace=True)
df['indrel_1mes'] = df['indrel_1mes'].astype(float).astype(np.int8)

features += ['age','antiguedad','renta','ind_nuevo','indrel','indrel_1mes','ind_actividad_cliente']
```

```python
df['fecha_alta_month'] = df['fecha_alta'].map(lambda x: 0.0 if x.__class__ is float else float(x.split('-')[1])).astype(np.int8)
df['fecha_alta_year'] = df['fecha_alta'].map(lambda x: 0.0 if x.__class__ is float else float(x.split('-')[0])).astype(np.int16)
features += ['fecha_alta_month', 'fecha_alta_year']

df['ult_fec_cli_1t_month'] = df['ult_fec_cli_1t'].map(lambda x: 0.0 if x.__class__ is float else float(x.split('-')[1])).astype(np.int8)
df['ult_fec_cli_1t_year'] = df['ult_fec_cli_1t'].map(lambda x: 0.0 if x.__class__ is float else float(x.split('-')[0])).astype(np.int16)
features += ['ult_fec_cli_1t_month', 'ult_fec_cli_1t_year']

df.fillna(-99, inplace=True)

def date_to_int(str_date):
    Y, M, D = [int(a) for a in str_date.strip().split("-")]
    int_date = (int(Y) - 2015)*12+int(M)
    return int_date

df['int_date'] = df['fecha_dato'].map(date_to_int).astype(np.int8)

df_lag = df.copy()
df_lag.columns = [col + '_prev' if col not in ['ncodpers', 'int_date'] else col for col in df.columns]
df_lag['int_date'] += 1

df_trn = df.merge(df_lag, on=['ncodpers','int_date'], how='left')

del df, df_lag

for prod in prods:
    prev = prod + '_prev'
    df_trn[prev].fillna(0, inplace=True)
df_trn.fillna(-99, inplace=True)

features+=[feature + '_prev' for feature in features]
features += [prod + '_prev' for prod in prods]
```

## XGBoost model

```python
import pandas as pd
import numpy as np
import xgboost as xgb

np.random.seed(2018)

# ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¨ë‹¤.
trn = pd.read_csv('kaggle/input/santander-product-recommendation-data/train_ver2.csv')
tst = pd.read_csv('kaggle/input/santander-product-recommendation-data/test_ver2.csv')

## ë°ì´í„° ì „ì²˜ë¦¬ ##

# ì œí’ˆ ë³€ìˆ˜ë¥¼ ë³„ë„ë¡œ ì €ìž¥í•´ ë†“ëŠ”ë‹¤.
prods = trn.columns[24:].tolist()

# ì œí’ˆ ë³€ìˆ˜ ê²°ì¸¡ê°’ì„ ë¯¸ë¦¬ 0ìœ¼ë¡œ ëŒ€ì²´í•œë‹¤.
trn[prods] = trn[prods].fillna(0.0).astype(np.int8)

# 24ê°œ ì œí’ˆ ì¤‘ í•˜ë‚˜ë„ ë³´ìœ í•˜ì§€ ì•ŠëŠ” ê³ ê° ë°ì´í„°ë¥¼ ì œê±°í•œë‹¤.
no_product = trn[prods].sum(axis=1) == 0
trn = trn[~no_product]

# í›ˆë ¨ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¥¼ í†µí•©í•œë‹¤. í…ŒìŠ¤íŠ¸ ë°ì´í„°ì— ì—†ëŠ” ì œí’ˆ ë³€ìˆ˜ëŠ” 0ìœ¼ë¡œ ì±„ìš´ë‹¤.
for col in trn.columns[24:]:
    tst[col] = 0
df = pd.concat([trn, tst], axis=0)

# í•™ìŠµì— ì‚¬ìš©í•  ë³€ìˆ˜ë¥¼ ë‹´ëŠ” listì´ë‹¤.
features = []

# ë²”ì£¼í˜• ë³€ìˆ˜ë¥¼ .factorize() í•¨ìˆ˜ë¥¼ í†µí•´ label encodingí•œë‹¤.
categorical_cols = ['ind_empleado', 'pais_residencia', 'sexo', 'tiprel_1mes', 'indresi', 'indext', 'conyuemp', 'canal_entrada', 'indfall', 'tipodom', 'nomprov', 'segmento']
for col in categorical_cols:
    df[col], _ = df[col].factorize(na_sentinel=-99)
features += categorical_cols

# ìˆ˜ì¹˜í˜• ë³€ìˆ˜ì˜ íŠ¹ì´ê°’ê³¼ ê²°ì¸¡ê°’ì„ -99ë¡œ ëŒ€ì²´í•˜ê³ , ì •ìˆ˜í˜•ìœ¼ë¡œ ë³€í™˜í•œë‹¤.
df['age'].replace(' NA', -99, inplace=True)
df['age'] = df['age'].astype(np.int8)

df['antiguedad'].replace('     NA', -99, inplace=True)
df['antiguedad'] = df['antiguedad'].astype(np.int8)

df['renta'].replace('         NA', -99, inplace=True)
df['renta'].fillna(-99, inplace=True)
df['renta'] = df['renta'].astype(float).astype(np.int8)

df['indrel_1mes'].replace('P', 5, inplace=True)
df['indrel_1mes'].fillna(-99, inplace=True)
df['indrel_1mes'] = df['indrel_1mes'].astype(float).astype(np.int8)

# í•™ìŠµì— ì‚¬ìš©í•  ìˆ˜ì¹˜í˜• ë³€ìˆ˜ë¥¼ featuresì— ì¶”êµ¬í•œë‹¤.
features += ['age','antiguedad','renta','ind_nuevo','indrel','indrel_1mes','ind_actividad_cliente']

# (í”¼ì³ ì—”ì§€ë‹ˆì–´ë§) ë‘ ë‚ ì§œ ë³€ìˆ˜ì—ì„œ ì—°ë„ì™€ ì›” ì •ë³´ë¥¼ ì¶”ì¶œí•œë‹¤.
df['fecha_alta_month'] = df['fecha_alta'].map(lambda x: 0.0 if x.__class__ is float else float(x.split('-')[1])).astype(np.int8)
df['fecha_alta_year'] = df['fecha_alta'].map(lambda x: 0.0 if x.__class__ is float else float(x.split('-')[0])).astype(np.int16)
features += ['fecha_alta_month', 'fecha_alta_year']

df['ult_fec_cli_1t_month'] = df['ult_fec_cli_1t'].map(lambda x: 0.0 if x.__class__ is float else float(x.split('-')[1])).astype(np.int8)
df['ult_fec_cli_1t_year'] = df['ult_fec_cli_1t'].map(lambda x: 0.0 if x.__class__ is float else float(x.split('-')[0])).astype(np.int16)
features += ['ult_fec_cli_1t_month', 'ult_fec_cli_1t_year']

# ê·¸ ì™¸ ë³€ìˆ˜ì˜ ê²°ì¸¡ê°’ì€ ëª¨ë‘ -99ë¡œ ëŒ€ì²´í•œë‹¤.
df.fillna(-99, inplace=True)

# (í”¼ì³ ì—”ì§€ë‹ˆì–´ë§) lag-1 ë°ì´í„°ë¥¼ ìƒì„±í•œë‹¤.
# ì½”ë“œ 2-12ì™€ ìœ ì‚¬í•œ ì½”ë“œ íë¦„ì´ë‹¤.

# ë‚ ì§œë¥¼ ìˆ«ìžë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜ì´ë‹¤. 2015-01-28ì€ 1, 2016-06-28ì€ 18ë¡œ ë³€í™˜ëœë‹¤
def date_to_int(str_date):
    Y, M, D = [int(a) for a in str_date.strip().split("-")] 
    int_date = (int(Y) - 2015) * 12 + int(M)
    return int_date

# ë‚ ì§œë¥¼ ìˆ«ìžë¡œ ë³€í™˜í•˜ì—¬ int_dateì— ì €ìž¥í•œë‹¤
df['int_date'] = df['fecha_dato'].map(date_to_int).astype(np.int8)

# ë°ì´í„°ë¥¼ ë³µì‚¬í•˜ê³ , int_date ë‚ ì§œì— 1ì„ ë”í•˜ì—¬ lagë¥¼ ìƒì„±í•œë‹¤. ë³€ìˆ˜ëª…ì— _prevë¥¼ ì¶”ê°€í•œë‹¤.
df_lag = df.copy()
df_lag.columns = [col + '_prev' if col not in ['ncodpers', 'int_date'] else col for col in df.columns ]
df_lag['int_date'] += 1

# ì›ë³¸ ë°ì´í„°ì™€ lag ë°ì´í„°ë¥¼ ncodperì™€ int_date ê¸°ì¤€ìœ¼ë¡œ í•©ì¹œë‹¤. Lag ë°ì´í„°ì˜ int_dateëŠ” 1 ë°€ë ¤ ìžˆê¸° ë•Œë¬¸ì—, ì €ë²ˆ ë‹¬ì˜ ì œí’ˆ ì •ë³´ê°€ ì‚½ìž…ëœë‹¤.
df_trn = df.merge(df_lag, on=['ncodpers','int_date'], how='left')

# ë©”ëª¨ë¦¬ íš¨ìœ¨ì„ ìœ„í•´ ë¶ˆí•„ìš”í•œ ë³€ìˆ˜ë¥¼ ë©”ëª¨ë¦¬ì—ì„œ ì œê±°í•œë‹¤
del df, df_lag

# ì €ë²ˆ ë‹¬ì˜ ì œí’ˆ ì •ë³´ê°€ ì¡´ìž¬í•˜ì§€ ì•Šì„ ê²½ìš°ë¥¼ ëŒ€ë¹„í•˜ì—¬ 0ìœ¼ë¡œ ëŒ€ì²´í•œë‹¤.
for prod in prods:
    prev = prod + '_prev'
    df_trn[prev].fillna(0, inplace=True)
df_trn.fillna(-99, inplace=True)

# lag-1 ë³€ìˆ˜ë¥¼ ì¶”ê°€í•œë‹¤.
features += [feature + '_prev' for feature in features]
features += [prod + '_prev' for prod in prods]

###
### Baseline ëª¨ë¸ ì´í›„, ë‹¤ì–‘í•œ í”¼ì³ ì—”ì§€ë‹ˆì–´ë§ì„ ì—¬ê¸°ì— ì¶”ê°€í•œë‹¤.
###


## ëª¨ë¸ í•™ìŠµ
# í•™ìŠµì„ ìœ„í•˜ì—¬ ë°ì´í„°ë¥¼ í›ˆë ¨, í…ŒìŠ¤íŠ¸ìš©ìœ¼ë¡œ ë¶„ë¦¬í•œë‹¤.
# í•™ìŠµì—ëŠ” 2016-01-28 ~ 2016-04-28 ë°ì´í„°ë§Œ ì‚¬ìš©í•˜ê³ , ê²€ì¦ì—ëŠ” 2016-05-28 ë°ì´í„°ë¥¼ ì‚¬ìš©í•œë‹¤.
use_dates = ['2016-01-28', '2016-02-28', '2016-03-28', '2016-04-28', '2016-05-28']
trn = df_trn[df_trn['fecha_dato'].isin(use_dates)]
tst = df_trn[df_trn['fecha_dato'] == '2016-06-28']
del df_trn

# í›ˆë ¨ ë°ì´í„°ì—ì„œ ì‹ ê·œ êµ¬ë§¤ ê±´ìˆ˜ë§Œ ì¶”ì¶œí•œë‹¤.
X = []
Y = []
for i, prod in enumerate(prods):
    prev = prod + '_prev'
    prX = trn[(trn[prod] == 1) & (trn[prev] == 0)]
    prY = np.zeros(prX.shape[0], dtype=np.int8) + i
    X.append(prX)
    Y.append(prY)
XY = pd.concat(X)
Y = np.hstack(Y)
XY['y'] = Y

# í›ˆë ¨, ê²€ì¦ ë°ì´í„°ë¡œ ë¶„ë¦¬í•œë‹¤. 
vld_date = '2016-05-28'
XY_trn = XY[XY['fecha_dato'] != vld_date]
XY_vld = XY[XY['fecha_dato'] == vld_date]


```

```python
# XGBoost ëª¨ë¸ parameterë¥¼ ì„¤ì •í•œë‹¤.
param = {
    'booster': 'gbtree',
    'max_depth': 8,
    'nthread': 4,
    'num_class': len(prods),
    'objective': 'multi:softprob',
    'silent': 1,
    'eval_metric': 'mlogloss',
    'eta': 0.1,
    'min_child_weight': 10,
    'colsample_bytree': 0.8,
    'colsample_bylevel': 0.9,
    'seed': 2018,
    }

# í›ˆë ¨, ê²€ì¦ ë°ì´í„°ë¥¼ XGBoost í˜•íƒœë¡œ ë³€í™˜í•œë‹¤.
#X_trn = XY_trn.as_matrix(columns=features)
#Y_trn = XY_trn.as_matrix(columns=['y'])
X_trn = XY_trn[features].values
Y_trn = XY_trn['y'].values
dtrn = xgb.DMatrix(X_trn, label=Y_trn, feature_names=features)

X_vld = XY_vld[features].values
Y_vld = XY_vld['y'].values
dvld = xgb.DMatrix(X_vld, label=Y_vld, feature_names=features)

# XGBoost ëª¨ë¸ì„ í›ˆë ¨ ë°ì´í„°ë¡œ í•™ìŠµí•œë‹¤!
watch_list = [(dtrn, 'train'), (dvld, 'eval')]
model = xgb.train(param, dtrn, num_boost_round=1000, evals=watch_list, early_stopping_rounds=20)

# í•™ìŠµí•œ ëª¨ë¸ì„ ì €ìž¥í•œë‹¤.
import pickle
pickle.dump(model, open("kaggle/input/santander-product-recommendation-data/xgb.baseline.pkl", "wb"))
best_ntree_limit = model.best_ntree_limit
```
