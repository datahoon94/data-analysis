## 밤 11시에 승차 인원이 가장 많은 지하철역은?

```python
import csv
f = open('../dataset/subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
mx = 0
mx_station = ''
t = int(input('몇 시일 때 승차 인원이 가장 많은 역이 궁금하세요? : '))

for row in data:
    row[4:] = map(int,row[4:])
    a = row[4+(t-4)*2]
    if a > mx:
        mx = a
        mx_station = row[3] +'('+row[1]+')'
print(mx_station ,mx)
```

## 지하철역 시간대별 승차 인원

```python
import csv
f = open('../dataset/subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
mx = [0] * 24
mx_station = [''] * 24
for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        a = row[j*2+4]
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3] + '(' + str((j+4)%24) + '시)'
import matplotlib.pyplot as plt
plt.rc('font', family="Malgun Gothic")
plt.bar(range(24),mx,color='r')
plt.xticks(range(24),mx_station, rotation=90)
plt.show()
```

## 지하철역 시간대별 하차 인원

```python
import csv
f = open('../dataset/subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
mx = [0] * 24
mx_station = [''] * 24
for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        b = row[j*2+5]
        if b > mx[j]:
            mx[j] = b
            mx_station[j] = row[3] + '(' + str((j+4)%24) + '시)'
import matplotlib.pyplot as plt
plt.rc('font', family="Malgun Gothic")
plt.bar(range(24),mx,color='b')
plt.xticks(range(24),mx_station, rotation=90)
plt.show()
```
